```python
import numpy as np 
import pandas as pd 
import re
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df_undersampled = pd.read_csv('https://raw.githubusercontent.com/bhanureddy01/CapstoneProject01/main/US_Accidents_March23_sampled_500k_Cleaned_v1.0.csv')
print("The shape of data is:",(df_undersampled.shape))
display(df_undersampled.head(3).T)
```

```python
X = df_undersampled.drop(['Severity', 'Severity_Merged'], axis=1)
y = df_undersampled['Severity_Merged']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1984)
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbtree = GradientBoostingClassifier(learning_rate=0.1, max_depth=1, random_state=1984)
gbtree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Predictions
y_pred_train = gbtree.predict(X_train)
y_pred_test = gbtree.predict(X_test)

# Calculate accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import classification_report

# Predictions
y_pred_train = gbtree.predict(X_train)
y_pred_test = gbtree.predict(X_test)

# Generate classification report
train_report = classification_report(y_train, y_pred_train)
test_report = classification_report(y_test, y_pred_test)

print("Training Classification Report:")
print(train_report)
print("\nTesting Classification Report:")
print(test_report)
```
